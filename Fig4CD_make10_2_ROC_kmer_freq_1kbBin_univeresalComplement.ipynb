# ROC

```python
def reverse_complement_seq(seq):
    complementary_dict = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C', 'N': 'N'}  
    return ''.join([complementary_dict[c] for c in seq[::-1]])
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score
```

## universal v.s. nonCHM

```python
os.chdir("/mnt/Storage/home/wangyiman/CHMsInOtherContexts/CellStateTransition/CHMOrganization/Universal_specific/Sequence/AUC_1kb")
kmer_freq_mat_uni = pd.read_csv(f'Universal_kmer_frequency.tsv', sep = "\t", header = 0)
kmer_freq_mat_non = pd.read_csv(f'NonCHMsCpGrich_kmer_frequency.tsv', sep = "\t", header = 0)
auc_dict_uni_non = {}
kmer_freq_mat_total = pd.concat([kmer_freq_mat_uni, kmer_freq_mat_non]).reset_index(drop = True).fillna(0)
kmer_freq_mat_total = kmer_freq_mat_total.loc[:, (kmer_freq_mat_total != 0).any(axis=0)]    
for kmer in kmer_freq_mat_total.columns[1:] :
    if reverse_complement_seq(kmer) in auc_dict_uni_non.keys() :
        continue
    else :
        df = kmer_freq_mat_total[['CHM_name', kmer]]
        label_str = df['CHM_name'].str.split('_', expand = True)[0]
        label_int = [1 if x == 'Universal' else 0 for x in label_str]
        df = df.fillna(0)
        auc_dict_uni_non[kmer] = roc_auc_score(label_int, df[kmer].tolist())
```

```python
### ZFP57 meth motif (universal v.s. nonCHM)
### zhanglab: make10_3_sequence_feature_ZFP57_methCpG_1kb.ipynb
zfp57_meth_countperkb_2of3 = pd.read_csv(
    '/mnt/Storage/home/wangyiman/CHMsInOtherContexts/CellStateTransition/CHMOrganization/Universal_specific/Sequence/ZFP57Frequency_1kb/methCpG/ZFP57methMotif_countPerKb_2of3.tsv',
    sep = "\t", header = 0, 
)
label_str = zfp57_meth_countperkb_2of3['name']
label_int = [1 if x == 'Universal' else 0 for x in label_str]
auc_dict_uni_non['TGCmCGC'] = roc_auc_score(label_int, (zfp57_meth_countperkb_2of3['if_meth']/1000).tolist())

auc_dict_uni_non['TGCmCGC']
```

```python
auc_df_uni_non = pd.DataFrame.from_dict(auc_dict_uni_non, orient='index').sort_values([0], ascending = False).reset_index()
auc_df_uni_non.columns = ['kmer', 'AUC']
auc_df_uni_non.to_csv('/mnt/Storage/home/wangyiman/CHMsInOtherContexts/CellStateTransition/CHMOrganization/Universal_specific/Sequence/AUC_1kb/AUC_Universal_vs_NonCHM_rmDup.tsv',
                       sep = "\t", index = False)
    
```

```python
auc_df_uni_non_bothSide = auc_df_uni_non.copy()
auc_df_uni_non_bothSide['opos'] = 1 - auc_df_uni_non_bothSide['AUC']
auc_df_uni_non_bothSide['bothside_AUC'] = auc_df_uni_non_bothSide[['AUC', 'opos']].max(axis = 1)
auc_df_uni_non_bothSide.sort_values(['bothside_AUC'], ascending= False).to_csv('/mnt/Storage/home/wangyiman/CHMsInOtherContexts/CellStateTransition/CHMOrganization/Universal_specific/Sequence/AUC_1kb/AUC_Universal_vs_NonCHM_rmDup_bothSide.tsv',
                       sep = "\t", index = False)
```
